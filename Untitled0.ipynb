```python
# --- 2) 読み込み設定（MATLABと同じ） ---
sheet_name = "Sheet44"

# x: C3:C23 (21点)  / y: D3:M23 (10列)
# pandas は 0-index のため、Excelの行3は index=2
df = pd.read_excel(xlsx_path, sheet_name=sheet_name, header=None, engine="openpyxl")

# C列 = index 2, 行3-23 = 2:23 (23は排他的でなくpandasはスライス終端含むので注意)
x = df.iloc[2:23, 2].to_numpy(dtype=float)

# D～M列 = index 3..12
Y = df.iloc[2:23, 3:13].to_numpy(dtype=float)  # shape (21,10)

x.shape, Y.shape
```

```python
# --- 3) fourier8 相当の関数定義 ---
# y = a0 + Σ_{k=1..8} [a_k cos(k*w*x) + b_k sin(k*w*x)]
def fourier8(x, a0,
             a1,b1,a2,b2,a3,b3,a4,b4,a5,b5,a6,b6,a7,b7,a8,b8,
             w):
    y = np.full_like(x, a0, dtype=float)
    coeffs = [(a1,b1),(a2,b2),(a3,b3),(a4,b4),(a5,b5),(a6,b6),(a7,b7),(a8,b8)]
    for k,(ak,bk) in enumerate(coeffs, start=1):
        y += ak*np.cos(k*w*x) + bk*np.sin(k*w*x)
    return y

# 初期値（ざっくり）：a0=0.5, 他0, wは 2π/周期 の目安
# x範囲に対して周期をざっくり 1周～2周想定してみる
xspan = x.max() - x.min()
w0 = 2*np.pi / xspan  # 1周期/範囲
p0 = [0.5] + [0.0]*16 + [w0]  # a0 + (a1,b1..a8,b8=16個) + w
len(p0)
```

```python
# --- 4) 交点（y=0.2,0.8）を「指定した近傍」から探す関数 ---
def find_root_near(func, target, x0, x_grid):
    """
    func: 連続関数 f(x)
    target: f(x)=target の解
    x0: 近傍中心
    x_grid: 評価用の細かいグリッド（単調である必要はないが、ここでは単調）
    """
    y = func(x_grid) - target

    # x0近傍を優先：|x - x0| が小さい順に符号変化区間を探す
    order = np.argsort(np.abs(x_grid - x0))
    # ただし order のままだと隣接点がバラバラなので、x順で区間を見る必要あり
    # → x0 付近のウィンドウを取って、その中で符号変化を探す
    # ウィンドウ幅（点数）を徐々に広げる
    for half in [20, 50, 100, 200, 400, 800, 1200]:
        # x0に近い点を half*2+1 個取り出し、x順に並べる
        idx = order[:min(len(x_grid), 2*half+1)]
        idx = np.sort(idx)
        xw = x_grid[idx]
        yw = y[idx]

        # 符号変化区間を探す
        sign = np.sign(yw)
        for i in range(len(xw)-1):
            if sign[i] == 0:
                return xw[i]
            if sign[i]*sign[i+1] < 0:
                # 線形補間で根
                x1,x2 = xw[i], xw[i+1]
                y1,y2 = yw[i], yw[i+1]
                return x1 - y1*(x2-x1)/(y2-y1)

    return np.nan
```

```python
# --- 5) 列ごとに処理（正規化→フィット→sol1..4→sharpness→描画） ---
targets = {
    "sol1": (0.2, 2.6),
    "sol2": (0.2, 6.3),
    "sol3": (0.8, 3.3),
    "sol4": (0.8, 6.0),
}

sharpness_values = []
fit_params_list = []

fig, axes = plt.subplots(2, 5, figsize=(18, 7), constrained_layout=True)

for col in range(10):
    y = Y[:, col].astype(float)

    # 正規化（MATLABと同じ式）
    y_min, y_max = np.nanmin(y), np.nanmax(y)
    y_norm = (y - y_min) / (y_max - y_min + 1e-12)

    # フィット（curve_fit）
    # w は正に制約したいので bounds を設定
    lower = [-np.inf] + [-np.inf]*16 + [0.0]
    upper = [ np.inf] + [ np.inf]*16 + [np.inf]
    popt, pcov = curve_fit(fourier8, x, y_norm, p0=p0, bounds=(lower, upper), maxfev=200000)

    fit_params_list.append(popt)

    # 連続関数化（細かいグリッドで評価）
    x_dense = np.linspace(x.min(), x.max(), 5000)
    f_dense = lambda xx: fourier8(xx, *popt)

    # MATLABの vpasolve の「初期値の近傍」っぽく、指定近傍で根を拾う
    sol = {}
    for name, (level, x0) in targets.items():
        sol[name] = find_root_near(f_dense, level, x0, x_dense)

    sol1, sol2, sol3, sol4 = sol["sol1"], sol["sol2"], sol["sol3"], sol["sol4"]

    # sharpness
    # MATLAB: 2/((sol3-sol1)+(sol2-sol4))
    denom = (sol3 - sol1) + (sol2 - sol4)
    sharpness = 2.0 / denom if np.isfinite(denom) and denom != 0 else np.nan
    sharpness_values.append(sharpness)

    # plot
    ax = axes[col//5, col%5]
    ax.plot(x, y_norm, "o", markersize=4)
    ax.plot(x_dense, f_dense(x_dense), "-", linewidth=1.5)
    ax.axhline(0.2, linestyle="-")
    ax.axhline(0.8, linestyle="-")

    # 解の点と垂線
    for key in ["sol1","sol2","sol3","sol4"]:
        sx = sol[key]
        if np.isfinite(sx):
            sy = f_dense(np.array([sx]))[0]
            ax.plot([sx], [sy], "o")
            ax.plot([sx, sx], [0, sy], "--", linewidth=1)

    # タイトル（D～M）
    col_letter = chr(ord("D") + col)
    ax.set_title(f"列 {col_letter}  sharpness={sharpness:.4g} mm^-1")
    ax.set_xlabel("x")
    ax.set_ylabel("normalized y")
    ax.set_aspect("equal", adjustable="box")
    ax.set_ylim(0, 1.05)

plt.show()

sharpness_values = np.array(sharpness_values, dtype=float)
sharpness_values
```
